```python
import csv
import pandas as pd
import csv
import pandas as pd
from glob import glob
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
f = open("paper-relative.csv")
```

```python
lines = [i for i in csv.reader(f)]
```

```python
data = [[int(j) for j in i[1:]] for i in lines[1:]]
```

```python
titles = [i[0] for i in lines[1:]]
```

```python
df = pd.DataFrame(data, columns=lines[0][1:], index=titles)
```

```python
? plt.colorbar
```

```python
plt.matshow(
    df,
    interpolation='nearest',
    vmax=20, vmin=-100
)
# xlabels = [i for i in df.columns]
xlabels = ["4.12", "4.15"]
ylabels = titles
plt.xticks(np.arange(len(xlabels)), xlabels, rotation='vertical')
plt.yticks(np.arange(len(ylabels)), ylabels)
plt.colorbar()

for (i, j), z in np.ndenumerate(df):
    plt.text(j, i, '{:0.0f}'.format(z), ha='center', va='center', 
#              color='w'
            )
```
